```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
column = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Class Labels"]
#load data
data = '/content/Iris.csv'
df = pd.read_csv(data, names=column)
df.head()
#num_lines = df.shape[0]
#print(num_lines)
```

```python
df.describe()
```

```python
sns.pairplot(df,hue="Class Labels")
```

Separating input column and output column

```python
data = df.values

x = data[:,0:4]
y = data[:,4]
#print(x) #input values
#print(y) #output values
```

Splitting Data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
# x_train = 120 inputData and y_train = 120 outputData
# x_test = 30 inputData and y_test = 30 outputData
print(y_test)
```

Model 1: Support Vector Machine Algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
prediction1 = model_svc.predict(x_test)
from sklearn.metrics import accuracy_score
#print(accuracy_score(y_test,prediction1)*100)
for i in range(len(prediction1)):
  #print(y_test[i],prediction1[i])
  if(y_test[i]==prediction1[i]):
    print("same")
  else: print("not same")
```

Model 2: Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(x_train,y_train)
```

```python
prediction2 = model_lr.predict(x_test)
from sklearn.metrics import accuracy_score
#print(accuracy_score(y_test,prediction2)*100)
for i in range(len(prediction2)):
#  print(y_test[i],prediction2[i])
  if(y_test[i]==prediction2[i]):
    print("same")
  else: print("not same")
```

Model 3: Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier()
model_dtc.fit(x_train,y_train)
```

```python
prediction3 = model_dtc.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction3)*100)
#for i in range(len(prediction3)):
  #print(y_test[i],prediction3[i])
#  if(y_test[i]==prediction3[i]):
#    print("same")
#  else: print("not same")
```

Classification report: It gives us a detailed summary of how our model is performing.

```python
from sklearn.metrics import classification_report
#print(classification_report(y_test,prediction1))
print(classification_report(y_test,prediction2))
#print(classification_report(y_test,prediction3))
```
